```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy
import openpyxl
from openpyxl import load_workbook
import os
import pickle
import datetime as dt
import seaborn as sns
import matplotlib.dates as mdates
from datetime import datetime
import matplotlib.patches as mpatches
```

```python
ox_df = pd.read_csv("Sample_time_windows_Oxygen, Dissolved as O2.csv")
temp_df = pd.read_csv("Sample_time_windows_Temperature of Water.csv")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 20))

for index, row in temp_df.iterrows():
    start = mdates.date2num(pd.to_datetime(row['Earliest date']))
    end = mdates.date2num(pd.to_datetime(row['Latest date']))
    #bar_height_factor = row['pH']/1000
    bar_height_factor = 0.4
    bar_color = 'red'

    # Define the start date
    start_date = datetime(1990, 1, 1)
    mid_date = datetime(2006, 1, 1)
    end_date = datetime(2022, 1, 1)
    # Convert the start date to a numerical format
    period_start = mdates.date2num(start_date)
    period_mid = mdates.date2num(mid_date)
    period_end = mdates.date2num(end_date)

    if end > period_end or start < period_start:
        bar_color = 'grey'

    if end > period_end and start < period_start:
        bar_color = 'black'

    axs[0].barh(str(row['Estuary']), end - start, left=start,height=bar_height_factor, color = bar_color)
    axs[0].axvline(x=period_start, color='black', linestyle='--', alpha=0.2)
    axs[0].axvline(x=period_mid, color='blue', linestyle='--', alpha=0.2)
    axs[0].axvline(x=period_end, color='black', linestyle='--', alpha=0.2)
    axs[0].tick_params(axis='y', labelsize=14)
    axs[0].tick_params(axis='x', labelsize=14)
    axs[0].set_ylabel("Estuary",fontsize = 16)
    axs[0].set_xlabel("Date",fontsize = 16)
    axs[0].text(0.01, 0.995, "A - Water Temperature", transform=axs[0].transAxes, fontsize=18,
             verticalalignment='top', horizontalalignment='left')


for index, row in ox_df.iterrows():
    start = mdates.date2num(pd.to_datetime(row['Earliest date']))
    end = mdates.date2num(pd.to_datetime(row['Latest date']))
    #bar_height_factor = row['pH']/1000
    bar_height_factor = 0.4
    bar_color = 'red'

    # Define the start date
    start_date = datetime(1990, 1, 1)
    mid_date = datetime(2006, 1, 1)
    end_date = datetime(2022, 1, 1)
    # Convert the start date to a numerical format
    period_start = mdates.date2num(start_date)
    period_mid = mdates.date2num(mid_date)
    period_end = mdates.date2num(end_date)

    if end > period_end or start < period_start:
        bar_color = 'grey'

    if end > period_end and start < period_start:
        bar_color = 'black'

    axs[1].barh(str(row['Estuary']), end - start, left=start,height=bar_height_factor, color = bar_color)
    axs[1].axvline(x=period_start, color='black', linestyle='--', alpha=0.2)
    axs[1].axvline(x=period_mid, color='blue', linestyle='--', alpha=0.2)
    axs[1].axvline(x=period_end, color='black', linestyle='--', alpha=0.2)
    axs[1].tick_params(axis='y', labelsize=14)
    axs[1].tick_params(axis='x', labelsize=14)
    axs[1].set_ylabel("Estuary",fontsize = 16)
    axs[1].set_xlabel("Date",fontsize = 16)
    axs[1].text(0.01, 0.995, "B - Dissolved Oxygen", transform=axs[1].transAxes, fontsize=18,
        verticalalignment='top', horizontalalignment='left')

# Configure date formatting for the x-axis
axs[0].xaxis_date()
axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axs[1].xaxis_date()
axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Create custom legend patches
red_patch = mpatches.Patch(color='red', label='Data does not cross any period')
grey_patch = mpatches.Patch(color='grey', label='Data crosses one period')
black_patch = mpatches.Patch(color='black', label='Data crosses all periods')

# Add legend to the plot
legend = axs[0].legend(handles=[red_patch, grey_patch, black_patch], loc='upper right', fontsize=12)
legend = axs[1].legend(handles=[red_patch, grey_patch, black_patch], loc='upper right', fontsize=12)

              
# Set x-axis label
axs[0].set_xlabel('Date')
axs[1].set_xlabel('Date')
axs[0].invert_yaxis()
axs[1].invert_yaxis()
fig.tight_layout()
plt.savefig('Time_periods_sampling_windows.png')
```
