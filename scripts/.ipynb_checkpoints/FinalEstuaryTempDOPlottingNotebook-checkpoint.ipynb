```python
# Create DataFrame
#df2 = pd.DataFrame(data_whole)
df = pd.DataFrame(data_t_m2_a)
df_do = pd.DataFrame(data_ox_m2_a)

# Drop rows with estuary 'coquet' and 'axe'
#data_model2_av = df[~df['Estuary'].isin(['Coquet', 'Axe'])]
#data_model2_do_av = df_do[~df_do['Estuary'].isin(['Coquet', 'Welland'])]
average_model_2 = df['Average'].mean()
average_model_2_do = df_do['Average'].mean()

# Sort DataFrame by 'Average' column in descending order
df = df.sort_values(by='Average', ascending=True)
df['Estuary'] = df['Estuary'].str.replace('_', ' ')
df['Estuary'] = df['Estuary'].replace('Taw Torridge', 'Taw torridge')

#custom_order = df['Estuary'] 
df_do = df_do.sort_values(by='Average', ascending=True)
df_do['Estuary'] = df_do['Estuary'].str.replace('_', ' ')
df_do['Estuary'] = df_do['Estuary'].replace('Taw Torridge', 'Taw torridge')

# Define colors based on sign of 'Average' values
colors = np.where(df['Average'] >= 0, 'darkred', 'darkblue')
# Define colors based on sign of 'Average' values
colors_do = np.where(df_do['Average'] >= 0, 'darkred', 'darkblue')


# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 16))

# First subplot
#axs[0].set_title('Annual temperature change between 1990 and 2022 - Model 2',fontsize = 16)
for i in range(len(df)):
    axs[0].errorbar(df['Average'].iloc[i], df['Estuary'].iloc[i], xerr=df['Std'].iloc[i], fmt='o', capsize=0, color=colors[i])
axs[0].axvline(x=0, color='black', linestyle='-')
axs[0].axvline(x=average_model_2, color='red', linestyle='--', alpha=0.5)
axs[0].set_xlabel('Average temperature change (C/year)',fontsize = 16)
axs[0].set_xlim([-0.2, 0.2])
axs[0].tick_params(axis='both', labelsize=14)
axs[0].set_ylabel('Estuary', fontsize = 16)
axs[0].grid(True)
bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="red", lw=1, alpha=0.5)
axs[0].text(0.185, -1, f'Average: {average_model_2:.3f} C/year', fontsize=12, color='red', ha='right', va='center', bbox=bbox_props)
axs[0].text(-0.197, 52.75, "A", fontsize=18, fontweight='bold')


# Second subplot
#axs[1].set_title('Annual dissolved oxygen change between 1990 and 2022 - Model 2', fontsize = 16)
# Add your second subplot code here
for i in range(len(df_do)):
    axs[1].errorbar(df_do['Average'].iloc[i], df_do['Estuary'].iloc[i], xerr=df_do['Std'].iloc[i], fmt='o', capsize=0, color=colors_do[i])
axs[1].axvline(x=0, color='black', linestyle='-')
axs[1].axvline(x=average_model_2_do, color='blue', linestyle='--', alpha=0.5)
axs[1].set_xlabel('Average dissolved oxygen change (mg/L/yr)',fontsize = 16)
axs[1].set_xlim([-0.2, 0.2])
axs[1].tick_params(axis='both', labelsize=14)
axs[1].set_ylabel('Estuary', fontsize = 16)
axs[1].grid(True)
bbox_props_do = dict(boxstyle="round,pad=0.3", fc="white", ec="blue", lw=1, alpha=0.5)
axs[1].text(0.165, -1, f'Average: {average_model_2_do:.3f} mg/L/year', fontsize=12, color='blue', ha='right', va='center', bbox=bbox_props_do)
axs[1].text(-0.197, 30.97, "B", fontsize=18, fontweight='bold')



# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig('Model_2_run_subplots_DO_temp.png')

plt.show()
```

```python
#Cumulative frequency plot
df_t_m4_y = pd.DataFrame(data_t_m4_y)
df_ox_m4_y = pd.DataFrame(data_ox_m4_y)

# Sort the DataFrame by the 'Average' column
sorted_df_t = df_t_m4_y.sort_values(by='Average')
sorted_df_ox = df_ox_m4_y.sort_values(by='Average')

# Calculate the cumulative sum of absolute frequencies
cumulative_freq_t = sorted_df_t['Average'].abs().cumsum()
cumulative_freq_ox = sorted_df_ox['Average'].abs().cumsum()

# Calculate the proportion of sites out of 1
proportion_t = cumulative_freq_t / cumulative_freq_t.iloc[-1]
proportion_ox = cumulative_freq_ox / cumulative_freq_ox.iloc[-1]

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot for temperature change
ax[0].plot(sorted_df_t['Average'], proportion_t, marker='s', linestyle='-', color='red')

# Set title and labels
ax[0].set_title('A', loc='left', fontweight='bold', fontsize=14)
ax[0].set_xlabel('Average temperature change (C/year)', fontsize=12)
ax[0].set_ylabel('Fraction of sampling points', fontsize=12)
ax[0].axvline(x=0, color='black', linestyle='--', linewidth=2)

# Set font size for axis labels and numbers
ax[0].tick_params(axis='both', which='major', labelsize=12)

# Calculate the fraction of sites where average temperature change is below 0
fraction_below_zero_t = len(sorted_df_t[sorted_df_t['Average'] < 0]) / len(sorted_df_t)
print(fraction_below_zero_t)
ax[0].grid(True)

# Plot for dissolved oxygen change
ax[1].plot(sorted_df_ox['Average'], proportion_ox, marker='s', linestyle='-', color='blue')
ax[1].axvline(x=0, color='black', linestyle='--', linewidth=2)

# Set title and labels
ax[1].set_title('B', loc='left', fontweight='bold', fontsize=14)
ax[1].set_xlabel('Average dissolved oxygen change (mg/L/year)', fontsize=12)
ax[1].set_ylabel('Fraction of sampling points', fontsize=12)

# Set font size for axis labels and numbers
ax[1].tick_params(axis='both', which='major', labelsize=12)

# Calculate the fraction of sites where average dissolved oxygen change is below 0
fraction_below_zero_ox = len(sorted_df_ox[sorted_df_ox['Average'] < 0]) / len(sorted_df_ox)
print(fraction_below_zero_ox)
ax[1].grid(True)

plt.tight_layout()

#plt.savefig('Frequency_counts_temp_do.png')
```

```python
from scipy.stats import f_oneway

det_x = 'Dissolved oxygen (mg/L)'

# Step 1: Exclude rows without numerical temperature of water
df_with_temp = merged_df_eed[pd.to_numeric(merged_df_eed[det_x], errors='coerce').notnull()]

# Step 2: Subset where geomorphology is "x", "y", or "z"
#geomorphology_x = df_with_temp[df_with_temp['geomorphology'] == 'Bar Built Estuary'][det_x]
#geomorphology_y = df_with_temp[df_with_temp['geomorphology'] == 'Ria'][det_x]
#geomorphology_z = df_with_temp[df_with_temp['geomorphology'] == 'Coastal Plain'][det_x]

tidal_x = df_with_temp[df_with_temp['tidal_range'] == 'Exceptionally large'][det_x]
tidal_y = df_with_temp[df_with_temp['tidal_range'] == 'Microtidal'][det_x]
tidal_z = df_with_temp[df_with_temp['tidal_range'] == 'Mesotidal'][det_x]
tidal_a = df_with_temp[df_with_temp['tidal_range'] == 'Macrotidal'][det_x]



# Perform one-way ANOVA
#f_statistic, p_value = f_oneway(geomorphology_x, geomorphology_y, geomorphology_z)
f_statistic, p_value = f_oneway(tidal_x, tidal_y, tidal_z,tidal_a)

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("The difference between the groups is statistically significant (reject null hypothesis)")
else:
    print("There is no significant difference between the groups (fail to reject null hypothesis)")

print(p_value)
```
